```python
import pandas as pd
from pathlib import Path
```

```python
file_path = '../output/separated'
files_paths = [str(x) for x in Path(file_path).glob("*.txt")]
print(files_paths)
```

```python
resumen = []

for path in files_paths:
  print(f'ANALISIS {path}')


  reader = open(path, "r+")
  lines = reader.readlines()

  solo_oraciones = list(filter(lambda line: line != '\n', lines))
  cantidad_oraciones = len(solo_oraciones)
  print(f'  Cantidad de oraciones: {cantidad_oraciones}')

  grouped_articles = ''.join(lines).split('\n\n')
  cantidad_articulos = len(grouped_articles)
  print(f'  Cantidad de articulos: {cantidad_articulos}')

  oraciones_por_articulo = round(cantidad_oraciones / cantidad_articulos, 2)
  print(f'  Promedio de oraciones por articulo: {oraciones_por_articulo}')

  caracteres_por_oracion = list(map(lambda sentence: len(sentence.replace('\n','')), solo_oraciones))
  df_caracteres_por_oracion = pd.DataFrame(caracteres_por_oracion).describe()
  promedio_caracteres_por_oracion = round(df_caracteres_por_oracion[0]['mean'], 2)
  print(f'  Promedio de caracteres por oracion: {promedio_caracteres_por_oracion}')
  print('  Caracteres por oracion')
  print(df_caracteres_por_oracion[0])

  palabras_por_oracion = list(map(lambda sentence: len(sentence.split()), solo_oraciones))
  df_palabras_por_oracion = pd.DataFrame(palabras_por_oracion).describe()
  promedio_palabras_por_oracion = round(df_palabras_por_oracion[0]['mean'], 2)
  print('  Palabras por oracion')
  print(f'  Promedio de palabras por oracion: {promedio_palabras_por_oracion}')
  print(df_palabras_por_oracion[0])

  resumen.append({
                  'archivo': path,
                  'cant_articulos': cantidad_articulos, 
                  'cantidad_oraciones': cantidad_oraciones,
                  'oraciones_por_articulo': oraciones_por_articulo,
                  'promedio_caracteres_por_oracion': promedio_caracteres_por_oracion,
                  'promedio_palabras_por_oracion': promedio_palabras_por_oracion
                  })

  print('')
  print('')
```

```python
df_resumen = pd.DataFrame(data=resumen)
df_resumen
```

```python
df_resumen.to_csv('salida_separateds.csv')
```
