```python
def sanitice_text(txt):

  # estoy interesado en obtener todos los caracteres que no sean:

  # 1. ASCII
  whiteList = '\x00-\x7F'

  # 2. Â¿ Â¡ tildes y enies
  whiteList += 'Â¿Â¡Ã€-Ã¿Ã±Ã‘'

  regex = r'[^{}]'.format(whiteList)
  patt = re.compile(regex)

  sanitice_char = lambda c: unidecode(c) if patt.match(c) else c
  
  return "".join([sanitice_char(c) for c in txt])
```

Cuando el beneficiario de un fideicomiso en garantiÌa es una institucioÌn de intermediacioÌn financiera, como en este caso el BROU, la ley autoriza que el Banco sea al mismo tiempo beneficiario -acreedor- y fiduciario. En cuanto a las ventajas del fideicomiso en garantiÌa, la doctrina senÌƒala su mayor flexibilidad y celeridad derivada de la ejecucioÌn extrajudicial de los bienes, lo que lo distingue de otros instrumentos de garantiÌa como la hipoteca.
