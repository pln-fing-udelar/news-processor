```python
import pandas as pd
from collections import Counter
```

```python
def assign_label(row):
  string = str(row["vote_1"]) + " " + str(row["vote_2"]) + " " + str(row["vote_3"])
  str_list = string.split()
  ct = Counter(str_list)
  return max(ct, key=ct.get)
```

```python

file_name_path = "./datasets/prensaUyUnaClase.csv"
df_1 = pd.read_csv(file_name_path, header=None, usecols=[0,2], names=["text", "label"])
df_1 = df_1[["label", "text"]]
df_1['label'] = df_1['label'].replace({'Neg':0, 'Pos':1, 'Neu':2})
df_1
```

```python
file_name_path = "./datasets/prensaUyVariasClases.csv"
df_2 = pd.read_csv(file_name_path, header=None, usecols=[0,2,3,4], names=["text","vote_1","vote_2","vote_3"])
df_2 = df_2[df_2['vote_1'] != "empate"]
df_2 = df_2.fillna("")
df_2["label"] = df_2.apply(lambda row: assign_label(row), axis="columns", result_type="expand")
df_2 = df_2[["label", "text"]]
df_2['label'] = df_2['label'].replace({'Neg':0, 'Pos':1, 'Neu':2})
df_2
```

```python
# Se unifican ambos datasets 
data = pd.concat([df_1, df_2], ignore_index=True)
data.to_csv('./datasets/dataset_final.csv', index = False)
data
```
